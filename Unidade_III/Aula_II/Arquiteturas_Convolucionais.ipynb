## Arquitetura de Redes Convolucionais

> Algoritmo de Aprendizado Profundo muito utilizado no tratamento de imagens.

> Realiza a atribuição de pesos e vieses em vários aspectos de uma imagem de entrada.

> Estes pesos permitem atribuir diferenças entre imagens, algo muito utilizado no setor médico.

> Tem um menor pré-processamento do que outros modelos também destinados a classificação.

> Na questão dos filtros, as ConvNets não os criam do zero, são treinados, assim como as características.

As camadas das ConvNets fazem o aprendizado de múltiplos filtros, e vários filtros são necessários para que todas as informações sejam extraídas da imagem tratada. Dentro destes modelos temos algumas camadas:

**- Camada de Pooling ou Agrupamento:** Usa a função de valor máximo para processamento de imagens, mas também pode ser feita pela função de média, ou soma.

**- Camada Dropout:** Desliga alguns neurônios de forma não ordenada para auto regulação, embora ocorra apenas na fase de treinamento.

**- Camada Flatten:** cria um array linear através do redimensionamento dos processos das camadas anteriores, oferecendo apenas uma única dimensão.

**- Camada Dense:** A rede neural é aplicada na sua forma fully conected, conde a dimensão de saída e a função par ativação são informadas.

> O ConvNet abusa do fato de que as imagens são coleções de valores expressos em pixels.

> Mas isso não permite que sua matriz seja "achatada", ou seja, passando de uma configuração de 3x3 pixels para 9x1. Isso não facilita a aplicação do perceptron;

> Se o modelo for aplicado em uma imagem de baixa complexidade, o processo de previsão de classes será facilitado, e possivelmente antevisto, apresentando uma precisão média.

> Neste caso, imagens complexas não apresentariam qualquer precisão.

> Desta forma, as ConvNets são capazes de capturar as dependências espaciais e até temporais das imagens, desde que os filtros corretos sejam aplicados.

> Nas ConvNets um melhor ajuste do conjunto imagem é feito, pois o modelo faz a redução da quantidade de parâmetros e do reuso dos pesos.

> Nas redes convolucionais, cada camada extrai informações específicas dos dados de entrada. A informação passa por cada camada da rede, com a saída da camada anterior fornecendo a entrada para a próxima camada da rede.

> A informação passa por cada camada da rede, com a saída da camada anterior fornecendo a entrada para a próxima camada da rede.

> Simula a nossa capacidade cognitiva e pode identificar rostos, pessoas e outros elementos presentes em uma imagem.

> As redes convolucionais podem, portanto, ser usadas, por exemplo, para classificar imagens ou detectar objetos em uma cena.
