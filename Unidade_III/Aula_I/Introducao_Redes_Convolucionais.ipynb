## Introdução e implementação do perceptron

- Criado entre as décadas de 1950 e 1960 por Frank Rosenblatt (cienteista).
- Este modelo permite uma melghor e mais clara percepção de como uma rede neural funciona, no seu aspecto matemático, sendo uma excelente introdução.
- Modelo matemático para diversas entradas e que produz uma única saída binária.

## Exemplo:

- Caso o modelo do Perceptron tenha três entradas, sendo x1, x2, x3.
- Para calcular a saída, basta aplicar uma simples regra de pesos, neste caso, w1, w2, ..., dentro do grupo dos números reais.
- A saída de um neurônio, neste modelo, seja 0 ou 1, ocorre por uma soma ponderada, este valor é considerado um limiar, threshold, em inglês.
- Atenção: tal como no caso dos pesos, o threshold é um número real e um parâmetro do neurônio.

- O Perceptron é um modelo capaz de tomar decisões quando comprova evidências.

> Exemplo:
- Em casos da necessidade de se tratar a informação sobre a previsão do tempo, devido a existência de algum evento;
- Digamos que a cidade vai receber um festival de música e os organizadores precisam antever se precisam de equipamentos e acessórios para o caso de ocorrer a chuva;
- A decisão, se positiva, vai depender da resposta a algumas perguntas, como:
    - O bom tempo deve permanecer todo o fim de semana do evento?
    - Existe a necessidade de oferecer guarda-chuvas?
    - O transporte público sofrerá algum impacto negativo?

Na matemática, o cálculo de uma convolução contínua é dado pela integral desse produto, desde que as funções sejam integráveis no intervalo em análise. Na computação, para simplificar, utiliza-se um cálculo discreto, que realiza o somatório do produto:

> (f * g)(k) = h(k)
 
> h(k) = ∑(f(j) * g(k-j))

Quando utilizamos a convolução em Deep Learning, no contexto de processamento de imagens, precisamos de dois somatórios, já que normalmente estamos lidando com duas dimensões – altura e largura. Observe os dois somatórios (sigmas) na equação a seguir:

> (f * g)(x,y) = ∑∑(f(i,j) * g(x-i,y-j))
