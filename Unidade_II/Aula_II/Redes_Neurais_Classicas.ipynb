## MLP (Multilayer Perceptron)

> Uma MLP consiste em pelo menos três camadas de neurônios: entrada, oculta e saída.

> Com excessão dos neurônios de entrada, cada neurônio usa uma função de ativação não linear.

```python
from PIL import Image

# Carregar a imagem
imagem = Image.open(r'D:\Users\Nayan Couto\Cloud Drive\Documentos\Arquivos PDF, PPT, DOC\Ciências de Dados - Anhanguera Ampli\Inteligência Artificial\Redes Neurais e Deep Learning\RNA_Deep_Learning\Unidade_II\Aula_II\MLP.png')

# Exibir a imagem
imagem
```

> O termo MLP não se refere em um único Perceptron que possui várias camadas.

> O MLP pode utilizar diversas funções de ativação aleatórias, que podem variar de camada para camada.

> Grande poder de generalização através do ajuste dos pesos durante a fase de treinamento.
