## Arquitetura das Redes Neurais Artificiais

> A forma como os neurônios estão dispostos nas camadas e o padrão de conexão entre eles define a sua arquitetura.

> Existem diferentes arquiteturas de Redes Neurais Artificiais, cada uma com o seu propósito definido.

> A escolha da arquitetura é uma das maiores dificuldades na utilização das Redes Neurais Artificiais.

## Tipos de Redes Neurais Artificiais

> Atualmente, as Redes Neurais mais difundidas são as FEEDFORWARD, CONVOLUCIONAIS, RECORRENTES, AUTOENCODERS e GENERATIVAS.

> Nas Redes Neurais Feedforwards, cada camada se conecta à próxima camada.
