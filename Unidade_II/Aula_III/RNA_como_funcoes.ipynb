## Hiperparâmetros

> Variáveis de controle definidas para o processo de aprendizado.

> As Redes Neurais possuem vários hiperparâmetros que precisam ser definidos:

>> Número máximo de épocas;

>> Tipo da função de ativação;

>> Função de custo;

>> Inicialização dos pesos;

>> Otimizador;

>> Taxa de aprendizado;

>> Tamanho do lote;

>> A própria arquitetura da rede, como o número de camadas ocultas e o número de neurônios presentes em cada camada;

> O número máximo de épocas representa o número de vezes que ocorre a atualização dos pesos na direção dos pesos iniciais.

> O tipo da função de ativação corresponde a qual transformação não linear será aplicada em cada neurônio.

> A função de custo descreve o tipo de função que será utilizada para medir o erro obtido pelas predições da rede durante o processo de treinamento.
